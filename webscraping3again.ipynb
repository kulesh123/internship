5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
def lat_long(city):
    driver=webdriver.Chrome()
    driver.get('https://www.google.com/maps')
    search=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
    search.send_keys(city)
    click=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button')
    click.click()
    time.sleep(3)
    url=driver.current_url
    lat_long=re.findall(r'@(.*),',url)
    print(lat_long)
    if len(lat_long):
        lat_long_list=lat_long[0].split(",")
        if len(lat_long_list)>=2:
            lat=lat_long_list[0]
            long=lat_long_list[1]
        print("Latitude = {}, Longitude = {}".format(lat,long))
```

8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video

```python
youtube=input("enter text")
driver=webdriver.Chrome()
driver.get("https://www.youtube.com/")
search=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/div/ytd-masthead/div[4]/div[2]/ytd-searchbox/form/div[1]/div[1]/input')
search.send_keys(youtube)
find=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/div/ytd-masthead/div[4]/div[2]/ytd-searchbox/button')
find.click()
youtubeurl=[]
urls=driver.find_elements(By.XPATH,'//script[@id="base-js"]')
for i in urls:
        
        youtubeurl.append(i.get_attribute('src'))
        print(youtubeurl)

time.sleep(3)
comment=[]
upvote=[]
times=[]
for i in youtubeurl:
    driver.get(i)
time.sleep(3)
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,1000)")
    comments=driver.find_elements(By.XPATH,'//yt-formatted-string[@id="content-text"]')
    for i in comments[0:500]:
        name=i.text
        comment.append(name)
    upvote=driver.find_elements(By.XPATH,'//span[@id="vote-count-middle"]')
    for i in upvote[0:500]:
        new=i.text
        upvote.append(new)
    tm=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in tm[0:500]:
        tim=i.text
        times.append(tim)
df=pd.DataFrame({"Comment":comment,"Upvote":upvote,"Times":time})
print(df)

    
```
